```python
# imports

import tifffile as tiff
import numpy as np
from pathlib import Path
import os
```

```python
data_dir_path = "/Users/zaca2954/academics/anabaena/anabaena_mcdb6440/movies/"


mask = tiff.imread(Path(data_dir_path, "2020.3.5_ana33047_minusn_0003_Crop_series1_cellMask_cropped.tif"))
crop = tiff.imread(
    Path(data_dir_path, "2020.3.5_ana33047_minusn_0003_Crop_cropped.tif")
)
```

```python

print(mask.shape)
print(crop.shape)
```

```python
for i in range(0, mask.shape[0]):
    mask_img = mask[i, :, :]
    crop_img_bf = crop[i, :, :]
    crop_img_fl = crop[i, :, :]

    basename = "2020.3.5_ana33047_minusn_0003_"

    tiff.imsave(
        Path(data_dir_path, basename + "mask_" + str(i) + ".tif"), mask_img
    )

    tiff.imsave(
        Path(data_dir_path, basename + "crop_bf_" + str(i) + ".tif"), crop_img_bf
    )

    tiff.imsave(
        Path(data_dir_path, basename + "crop_fl_" + str(i) + ".tif"), crop_img_fl
    )
```

```python
train_data_path = 'fov_images/training_data'
```

```python
fov_train = glob.glob('fov_images/20241010_ZMB_Anabaena/20241010_001_ZMB*/tif/*_seg.npy')
fov_train
```

```python
for segfile in fov_train:
	brightfield = segfile.replace('_seg.npy', '.tiff')
	assert os.path.exists(brightfield)
	img = tifffile.imread(brightfield)

	try:
		seg = np.load(segfile,allow_pickle=True).item()
	except:
		print(segfile + ' failed to load')
		continue
	assert 'masks' in list(seg.keys())
	masks = seg['masks']

	image_set = os.path.basename(os.path.dirname(os.path.dirname(brightfield))).split('_')[-1]
	
	# save brightfield with "_img" and mask as "_labeled"
	imgfile = os.path.join(train_data_path, image_set+'.'+os.path.basename(brightfield).replace('.tiff', '_img.tiff'))
	maskfile = os.path.join(train_data_path, image_set+'.'+os.path.basename(brightfield).replace('.tiff', '_labeled.tiff'))
	tifffile.imwrite(imgfile, img)
	tifffile.imwrite(maskfile, masks)
```
